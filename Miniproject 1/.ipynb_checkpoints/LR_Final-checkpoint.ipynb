```python
import numpy as np
from random import shuffle
import matplotlib.pyplot as plt
import csv
import time 
```

```python
#Getting the data from the wine quality CSV file and the Cancer data file

# wine_matrix = np.matrix(['Fixed_Acidity', 'Volatile_Acidity', 'Citric_Acid', 'Residual_Sugar', 'Chlorides',
#              'Free_Sulfur_Dioxide', 'Total_Sulfur_Dioxide', 'Density', 'PH', 'Sulphates', 'Alcohol', 'Quality'])

wine_raw_data = []
with open('winequality-red.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    for index, item in enumerate(reader):
        if (index == 0):
            continue
        else:
            individual_row = np.array(item[0].split(';'))
            #Converting to binary. Cleaning up the data. 
            if ( int(individual_row[11]) >= 6):
                individual_row[11] = 1
            else:
                individual_row[11] = 0
                
            wine_raw_data.append(list(map(float, individual_row)))
            
csvFile.close()
shuffle(wine_raw_data)
wine_features = []
wine_labels = []

for row in wine_raw_data:
    wine_labels.append(row[len(row)-1])
    wine_features.append(row[0:len(row)-1])

wine_features = np.asarray(wine_features)
wine_labels = np.asarray(wine_labels)

# cancer_matrix = np.matrix(['Sample Code Number', 'Clump Thickness', 'U Cell Size', 'U Cell Shape',
#                          'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin',
#                          'Normal Nucleoli', 'Mitoses', 'Class'])

cancer_raw_data = []

with open('breast-cancer-wisconsin.data', 'r') as dataFile:
    cancer_reader = csv.reader(dataFile)
    for row in cancer_reader:
        #Removing bad data
        if '?' in row:
            continue
        else:
            cancer_raw_data.append(list(map(int, row)))
dataFile.close()
shuffle(cancer_raw_data)

cancer_features = []
cancer_labels = []

for row in cancer_raw_data:
    cancer_labels.append(row[len(row)-1])
    cancer_features.append(row[1:len(row)-1])

cancer_features = np.asarray(cancer_features)
cancer_labels = np.asarray(cancer_labels)

for index, element in np.ndenumerate(cancer_labels):
    if (element == 2):
        #0 for benign
        cancer_labels[index] = 0
    else:
        #1 for malignant
        cancer_labels[index] = 1
        
########################################Adding new Feature#########################################################

#Adding a new feature to the wine dataset
wine_features = np.c_[ wine_features, np.ones(len(wine_features)) ]

#Taking the logarithm of the product of the highest weight features
for y in range(len(wine_features)):
    log = np.log2(wine_features[y][0] * wine_features[y][5] * wine_features[y][10])
    #log = np.log2(wine_features[y][6])
    wine_features[y][11] = log

########################################Implementing Logistic Regression Here######################################

correction = 1e-5    
#Logistic function
def sigmoid(x):
    return 1/(1 + np.exp(-x))

#Weighted Sum of inputs. Dot product
def dot_product(theta, x):
    return np.dot(x, theta)

#Returns the probability after passing through sigmoid. Inputs needed are weights and x. 
def probability(theta, x):
    return sigmoid(dot_product(theta, x))

#Cost Function
#Our goal is to minimize this cost function by changing values of theta, i.e the weights
def cost_function(theta, x, y):
    #Get training input size
    m = x.shape[0]
    total_cost = -(1/m) * np.sum(
        y * np.log(probability(theta, x) + correction) + (1 - y) * np.log(
            1 - probability(theta, x) + correction))
    return total_cost

#This is the gradient of the above cost function at point theta. 
def gradient(theta, x, y):
    m = x.shape[0]
    return (1 / m) * np.dot(x.T, sigmoid(dot_product(theta,   x)) - y)

def new_fit(x, y, lr, iterations, threshold): #Imagine theta as a point 
    #Initializing random weights
    theta = np.zeros([x.shape[1]])
    return_theta = theta
    #Calculating the cost at that weight theta
    new_loss = cost_function(theta, x, y) + 1000
    #print (new_loss)
    
    for i in range (iterations):
        gradient1 = gradient(theta, x, y)
        #New theta = Old theta - (Step Size)
        theta = theta - (lr * gradient1)
        #Calculating the cost at the new weight theta
        loss = cost_function(theta, x, y)
        
        #Extra feature implemented here
        if (new_loss < threshold):
            break
            
        if (loss < new_loss):
            new_loss = loss
            return_theta = theta
        else:
            continue
    
    return return_theta       

def accuracy(params, x, actual_classes, probab_threshold=0.5):
    predicted_classes = (probability(params[:, np.newaxis], x) >= 
                         probab_threshold).astype(int)
    predicted_classes = predicted_classes.flatten()
    accuracy = np.mean(predicted_classes == actual_classes)
    return accuracy * 100
    
####################################Implementing 5 or k-fold cross validation here######################################

#This method returns the training data sets.
#Input is the index of the training set you want. The first, the second etc. Size is 136 or 319.  
def training_data_set(n, size, features, labels):
    #index_array contains all the indexes of the elements we want to remove
    index_array = []
    first_element_index = n*size
    for x in range(size):
        index_array.append(first_element_index + x)
        
    X_training_set = features
    #Adding 1 for the dummy column
    X_training_set = np.c_[ X_training_set, np.ones(len(features)) ]
    X_filtered_set = []

    Y_training_set = labels
    Y_filtered_set = []
    
    for y in range(len(features)):
        if y in index_array:
            continue
        else:
            X_filtered_set.append(X_training_set[y])
            Y_filtered_set.append(Y_training_set[y])
    
    #X_Filtered_set and Y_Filtered_set contain the training examples we want.
    #That is the training data, not the validation data
    return X_filtered_set, Y_filtered_set

########################################Testing for Accuracy#######################################################

def accuracy_final(size, features, labels, lr, iterations):
    #time2 = 0
    accuracy_final_value = 0
    for x in range(number_of_folds):
        Training_fold = training_data_set(x, size, features, labels)
        X_Training = Training_fold[0]
        Y_Training = Training_fold[1]

        X_Training = np.asarray(X_Training)
        Y_Training = np.asarray(Y_Training)
        
        #Threshold is the extra feature
        params = new_fit(X_Training, Y_Training, lr, iterations, 0.0001)
                                                
        X_test = features[x*size:x*size + size]
        X_test = np.c_[ X_test, np.ones(size) ]
        Y_test = labels[x*size:x*size + size]
        
        #time1 = time.time()
        accuracy_final_value += accuracy(params, X_test, Y_test.flatten())
        #time2 += (time.time() - time1)
    
    #print ((time2/number_of_folds)/size)
    return (accuracy_final_value/number_of_folds)

number_of_folds = 5
cancer_accuracy = accuracy_final(len(cancer_features)//number_of_folds, cancer_features, cancer_labels, 1, 10000)
wine_accuracy = accuracy_final(len(wine_features)//number_of_folds, wine_features, wine_labels, 1, 100000)

print (cancer_accuracy)
print (wine_accuracy)
```
